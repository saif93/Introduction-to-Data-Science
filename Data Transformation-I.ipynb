```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
wheat = pd.read_csv('wheat.data')
```

```python
iris = pd.read_csv(
    'https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
```

```python
wheat.head()
```

```python
wheat.drop(wheat.columns[0], axis=1, inplace=True)
```

```python
pd.plotting.scatter_matrix(wheat, figsize=(11,11))
```

```python
wheat.cov()
```

```python
plt.imshow(wheat.corr(), cmap=plt.cm.Blues, interpolation='nearest')
plt.colorbar()
tick_marks = [i for i in range(len(wheat.columns))]
plt.xticks(tick_marks, wheat.columns, rotation='vertical')
plt.yticks(tick_marks, wheat.columns)

plt.show()
```

# Corelation matrix using imshow

```python
students[students.columns[20:]].corr()
```

```python
plt.imshow(students[students.columns[20:]].corr(), cmap=plt.cm.Blues, interpolation='nearest')
plt.colorbar()
tick_marks = [i for i in range(len(students.columns[20:]))]
plt.xticks(tick_marks, students.columns[20:], rotation='vertical')
plt.yticks(tick_marks, students.columns[20:])

plt.show()
```

```python
plt.imshow(students.corr(), cmap=plt.cm.Blues, interpolation='nearest')
plt.colorbar()
tick_marks = [i for i in range(len(students.columns))]
plt.xticks(tick_marks, students.columns, rotation='vertical')
plt.yticks(tick_marks, students.columns)

plt.show()
```
