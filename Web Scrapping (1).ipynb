```python
#creating dataframe
import pandas as pd
data = pd.DataFrame(columns=['Product_ID', 'PName', 'Price','Image','URL'])
```

```python
for i in divs:
    id = re.search('\d+',i.get('class')[-1]).group()
    name = i.find_all('h5', class_='ProductDetails')[0].select('a')[0].get_text()
    img = i.find_all('img', class_='img-responsive')[0].get("data-src")
    url = i.select('a')[0].get("href")
    price = i.find_all('div', class_='ActualPrice')[0].get_text()
    print(price)
    price= re.search('(\d+),*(\d+)*',price).group()
    if len(price.split(','))>1:
        price = price.split(',')[0]+price.split(',')[1]
    else:
        price = price
    
    #appending to the dataframe
    data = data.append({'Product_ID': id, 'PName': name, 'Price': price, 'Image':img, 'URL':url}, ignore_index=True)

    
```
